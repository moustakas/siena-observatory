```python
import os
import numpy as np
from glob import glob
```

```python
from ccdproc import ImageFileCollection
from astropy.nddata import CCDData
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rawdata_dir(night=''):
    return os.path.join(os.getenv('BREYO_DATA_DIR'), 'raw', night)
```

```python
night = '20190323'
rawdir = rawdata_dir(night)
print(rawdir)
```

```python
ic = ImageFileCollection(rawdir, keywords='*', glob_include='*.fit*')
ic.summary[:3]
```

```python
CCDData.read(ic.files[0])
```
